## Base Practice

#### 1. Create a null vector of size 10 but the fifth value which is 1 (base)
`hint: array[4]`

```python
null_vec = np.zeros(10)
null_vec[4] = 1

print(null_vec)
```

#### 2. Create a 3x3 matrix with values ranging from 0 to 8 with zeros on the diagonal (base)
`hint: np.arange, reshape, np.eye`

```python
range_mat = np.arange(9).reshape(3, 3)
eye_mat = np.eye(3)
eyed_mat = range_mat - eye_mat * range_mat

print(eyed_mat)
```

#### 3. How to add a border (filled with 0's) around an 5x5 matrix of ones? (base)
`hint: np.ones, np.pad`

```python
ones_mat = np.ones((5,5))
padded_mat = np.pad(ones_mat, ((1,1), (1,1)), 'constant', constant_values=(0))

print(padded_mat)
```

#### 4. Normalize a 5x5 random matrix (base)
`hint: (x - mean) / std, np.mean, np.std`

```python
rand_mat = np.random.random((5,5))
norm_mat = (rand_mat - rand_mat.mean()) / rand_mat.std()

print(norm_mat)
```

#### 5. Multiply a 5x3 matrix of ones by a 3x2 matrix of ones (real matrix product) (base)
`hint: np.dot`

```python
dot_mat = np.dot(np.ones((5,3)), np.ones((3,2)))

print(dot_mat)
```

## Intermediate Practice

#### 6. Create a vector of size 10 with values ranging from 0 to 1, both excluded (intermediate)
`hint: np.linspace`

```python
lin_vec = np.linspace(0, 1, 10 + 1, endpoint=False)[1:]

print(lin_vec)
```

#### 7. Create a random vector of size 10 with values in range (-3, 12) and sort it (intermediate)
`hint: sort`

```python
sorted_vec = np.sort(np.random.uniform(-3, 12, 10))

print(sorted_vec)
```

#### 8. Create random vector of size 10 and replace the maximum value by its additive inverse and the minimum value (of the original array) with the median value (intermediate)
`hint: argmax, argmin`

```python
# Create random vector
arg_vec = np.random.random(10)

# Get max and min indices, and corresponding inverse and median
i_max, i_min = np.argmax(arg_vec), np.argmin(arg_vec)
inv_max = -arg_vec.max()
median = np.median(arg_vec)

# Replace within the original vector
arg_vec[i_max] = inv_max
arg_vec[i_min] = median

print(arg_vec)
```

#### 9. Randomly replace p elements in a 2D nxn zero matrix to 1 (intermediate)
`hint: np.put, np.random.choice`

```python
n = 5
p = 10
zero_mat = np.zeros((n,n))

zero_mat.put(np.random.choice(
    zero_mat.size,
    p,
    replace=False
), 1)

print(zero_mat)
```

#### 10. Subtract the mean of each row of a random 5x10 matrix (intermediate)
`hint: mean(axis=,keepdims=)`

```python
mean_mat = np.random.random((5, 10))
mean_mat -= mean_mat.mean(axis=1, keepdims=True)

print(mean_mat)
```
