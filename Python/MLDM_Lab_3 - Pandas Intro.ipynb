### Plot in pie charts the number of "tamponi_test_molecolare" with percentage for each region with exploded slice for *Lombardia* region

`hint: df.index, df[].plot.pie, figsize=, autopct="%1.1f%%, pctdistance=, explode=, ylabel=", `

- Figure size of (8,8)
- Explode of 0.1 for Lombardia region
- Distance of the percentage of 0.75

```python
covid_regions_df['tamponi_test_molecolare'].plot.pie(
    figsize=(8,8),
    autopct='%1.1f%%',
    pctdistance=0.75,
    explode=[0.15 if region == 'Lombardia' else 0.0 for region in covid_regions_df.index],
    ylabel="Tamponi Test Molecolare"
)
```

### Plot hexbin plot of "deceduti" by ("long", "lat) coordinates of regions

`hint: df.plot.hexbin, x=, y=, C=, edgecolor=, gridsize=`

### Plot scatterplots of "terapia_intensiva" and "ingressi_terapia_intensiva" by ("long", "lat") coordinates of regions both in the same plot

`hint: df.plot.scatter, x=, y=, color=, alpha=, s=, label=, ax=, ax.legend`

- Figure size of (8, 6)
- Alpha of 0.5
- Legend at the bottom

```python
zoom_factor = 50 # Very small values, so... big zoom

ax = covid_regions_df.plot.scatter(
    x='long',
    y='lat',
    s=covid_regions_df['terapia_intensiva'] * zoom_factor,
    c='violet',
    alpha=0.5,
    label='Terapia Intensiva'
)

covid_regions_df.plot.scatter(
    x='long',
    y='lat',
    s=covid_regions_df['ingressi_terapia_intensiva'] * zoom_factor,
    c='green',
    figsize=(8,6),
    alpha=0.5,
    label='Ingressi Terapia Intensiva',
    ax=ax
)

ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=2)
```

```python
regions_colors = {
    "Abruzzo": "orange",
    "Basilicata": "orange",
    "Calabria": "orange",
    "Campania": "red",
    "Emilia-Romagna": "red",
    "Friuli Venezia Giulia": "red",
    "Lazio": "red",
    "Liguria": "orange",
    "Lombardia": "red",
    "Marche": "red",
    "Molise": "red",
    "P.A. Bolzano": "orange",
    "P.A. Trento": "red",
    "Piemonte": "red",
    "Puglia": "red",
    "Sardegna": "white",
    "Sicilia": "red",
    "Toscana": "orange",
    "Umbria": "orange",
    "Valle d'Aosta": "orange",
    "Veneto": "red",
}

mean_new_positives_df = covid_regions_df['nuovi_positivi'].groupby(pd.Series(regions_colors)).mean()

mean_new_positives_df.plot.bar(
    figsize=(8,6),
    title='Media nuovi positivi per colore regione',
    rot=0,
    color=["gray" if index == "white" else index for index in mean_new_positives_df.index]
)
```

### Group the regions by color and plot bars of mean *and error* of "nuovi_positivi" column

`hint: df.groupby().std(), yerr=, rot=,`

```python
yerr = covid_regions_df['nuovi_positivi'].groupby(pd.Series(regions_colors)).std() # Standard deviation
covid_regions_df['nuovi_positivi'].groupby(pd.Series(regions_colors)).mean().plot.bar(
    figsize=(8,6),
    title='Media nuovi positivi per colore regione',
    yerr=yerr,
    rot=0,
    color=["gray" if index == "white" else index for index in mean_new_positives_df.index]
)
```

### Group the regions by color and plot a hist of "variazione_totale_positivi" using 5 bins

`hint: pd.Series, df.groupby().plot.hist(), legend=,`

```python
delta_tot_positives_df = covid_regions_df['variazione_totale_positivi'].groupby(pd.Series(regions_colors))

delta_tot_positives_df.plot.hist(
    bins=5,
    figsize=(8,6),
    title='Variazione totale positivi per colore regione',
    alpha = 0.5,
    rot = 0,
    legend=True
)
```
