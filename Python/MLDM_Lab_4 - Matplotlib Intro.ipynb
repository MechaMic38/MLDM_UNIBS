## Base

### Plot a line with formula $y = 2x + 1$
`hint: np.arange, plt.figure, plt.plot, plt.title, plt.xlabel, plt.ylabel, plt.grid, plt.legend, plt.show`

- Use points in range [1, 20]
- Figure size of (10,5)
- Set axis labels
- Set plot title
- Set line color as red
- Discontinued line (--) with star (\*) on point
- Plot the legend with the formula in latex version (`r"$ ... $"`)
- Set a dashed grid

```python
# Define points of the line
x = np.arange(1, 21, 1)
y_line = x * 2 + 1

# Plot the line
fig, ax = plt.subplots(
    figsize=(10, 5),
    layout='constrained'
)
ax.plot(x, y_line, "r*--", label=r"$y = 2x + 1$")
ax.set_title("Line")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(linestyle="--")
ax.legend()

# Display figure
plt.show()
```

### Add to the previous plot the line with formula $y = log(x) + 1$ in the same range
`hint: np.log, plt.plot`

- Set line color as blue
- Normal line (-) with square (s) char on point

```python
# Define points of the line
y_log = np.log(x) + 1

# Plot the additional line
fig, ax = plt.subplots(
    figsize=(10, 5),
    layout='constrained'
)
ax.plot(x, y_line, "r*--", label=r"$y = 2x + 1$")
ax.plot(x, y_log, "gs-", label=r"$y = log(x) + 1$")
ax.set_title("Line and logarithmic functions")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(linestyle="--")
ax.legend()

# Display figure
plt.show()
```

### Add to the previous plot the line with formula $y=x^2 - 1$ in the same range
`hint: x**2, plt.plot`

- Set line color as green
- Normal line (-) with + (plus) char on point

```python
# Define points of the line
y_parabola = x ** 2 - 1

# Plot the additional parabola
fig, ax = plt.subplots(
    figsize=(10, 5),
    layout='constrained'
)
ax.plot(x, y_line, "r*--", label=r"$y = 2x + 1$")
ax.plot(x, y_log, "gs-", label=r"$y = log(x) + 1$")
ax.plot(x, y_parabola, "b+-", label=r"$y = x^2 - 1$")
ax.set_title("Line, logarithmic and parabolic functions")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(linestyle="--")
ax.legend()

# Display figure
plt.show()
```

### Replot the previous plot in different subplots in the same line without sharing axes
`hint: plt.subplots, axes[].plot, axes[].set_title, axes[].set_xlabel, axes[].set_ylabel, axes[].grid, axes[].legend, fig.tight_layout`

```python
# Create subplots
fig, axs = plt.subplots(
    nrows=1,
    ncols=3,
    figsize=(10, 5),
    layout='tight'
)

axs[0].plot(x, y_line, "r*--", label=r"$y = 2x + 1$")
axs[0].set_title("Line")

axs[1].plot(x, y_log, "gs-", label=r"$y = log(x) + 1$")
axs[1].set_title("Logarithm")

axs[2].plot(x, y_parabola, "b+-", label=r"$y = x^2 - 1$")
axs[2].set_title("Parabola")

for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(linestyle="--")
    ax.legend()

# Display figure
plt.show()
```

### Replot the previous plot sharing y between suplots
`hint: sharey=True`

```python
# Create subplots
fig, axs = plt.subplots(
    nrows=1,
    ncols=3,
    sharey=True,
    figsize=(10, 5),
    layout='tight'
)

axs[0].plot(x, y_line, "r*--", label=r"$y = 2x + 1$")
axs[0].set_title("Line")

axs[1].plot(x, y_log, "gs-", label=r"$y = log(x) + 1$")
axs[1].set_title("Logarithm")

axs[2].plot(x, y_parabola, "b+-", label=r"$y = x^2 - 1$")
axs[2].set_title("Parabola")

for ax in axs:
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(linestyle="--")
    ax.legend()

# Display figure
plt.show()
```

### Replot the previous plot using the for loop, lists and dictionaries
`hint: functions = [...], {"title": ..., "y": ..., "label": ..., "linestyle": ...}, zip(functions, axes)`

```python
# Define functions dictionary
functions = [
    {"title": "Line", "y": y_line, "label": r"$y = 2x + 1$", "linestyle": "r*--"},
    {"title": "Logarithm", "y": y_log, "label": r"$y = log(x) + 1$", "linestyle": "gs-"},
    {"title": "Parabola", "y": y_parabola, "label": r"$y = x^2 - 1$", "linestyle": "b+-"}
]

# Create subplots
fig, axs = plt.subplots(
    nrows=1,
    ncols=3,
    sharey=True,
    figsize=(10, 5),
    layout='tight'
)

for fn, ax in zip(functions, axs):
    ax.plot(x, fn["y"], fn["linestyle"], label=fn["label"])
    ax.set_title(fn["title"])
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.grid(linestyle="--")
    ax.legend()

# Display figure
plt.show()
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in the same plot with size (10,5) and a different color for each function

```python
# Plot functions
fig, ax = plt.subplots(
    figsize=(10, 5),
    layout='constrained'
)

for i in range(10):
    ax.plot(x, x * i - 5*i, label=fr"$y_{i} = {i}x - {5*i}$")

ax.set_title("Functions " + r"$y_i = i \cdot x - 5i$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.grid(linestyle="--")
ax.legend()

# Display figure
plt.show()
```

### Plot functions $ y_i = i\cdot x - 5i$ with $i$ in range (0, 10) in different subplots with shared y and figure size (20,5)

```python
# Plot functions
fig, axs = plt.subplots(
    nrows=2,
    ncols=5,
    sharey=True,
    figsize=(20, 5),
    layout='constrained'
)
axs = axs.flatten() # Work with 1D array instead of 2D matrix

for i in range(10):
    axs[i].plot(x, x * i - 5*i, label=fr"$y_{i} = {i}x - {5*i}$")
    axs[i].set_xlabel("x")
    axs[i].set_ylabel("y")
    axs[i].grid(linestyle="--")
    axs[i].legend()

fig.suptitle("Functions " + r"$y_i = i \cdot x - 5i$")

# Display figure
plt.show()
```

## Intermediate

### Plot a red scatterplot of $x$ squared
`hint: plt.scatter`

```python
fig = plt.figure(
    figsize=(10, 5),
    layout='constrained'
)

plt.scatter(x, x**2, c="red", label=r"$y = x^2$")
plt.title("Scatter Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle="--")
plt.legend()

plt.show()
```

### Plot a blue step plot of $x$ squared with linewidth of $3$
`hint: plt.step`

```python
fig = plt.figure(
    figsize=(10, 5),
    layout='constrained'
)

plt.step(x, x**2, c="blue", linewidth=3, label=r"$y = x^2$")
plt.title("Step Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle="--")
plt.legend()

plt.show()
```

### Plot an orange barplot plot of $x$ squared with alpha of $0.25$
`hint: plt.bar`

```python
fig = plt.figure(
    figsize=(10, 5),
    layout='constrained'
)

plt.bar(x, x**2, color="orange", alpha=0.25, label=r"$y = x^2$")
plt.title("Bar Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle="--")
plt.legend()

plt.show()
```

### Plot the area between $x$ squared and the $x \cdot log(x)$ function in green and alpha $0.5$
`hint: plt.fill_between`

```python
fig = plt.figure(
    figsize=(10, 5),
    layout='constrained'
)

plt.fill_between(x, x**2, x * np.log(x), color="green", alpha=0.5, label="Area = " + r"$x^2 - x \log(x)$")
plt.title("Area Between Two Functions Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle="--")
plt.legend(loc=2)

plt.show()
```

### Plot purple histogram of 50 bins and pink cumulative instogram of 100K random samples in subplots
`hint: np.random.randn, axes[].hist(), cumulative=True, bins=50`

```python
# Random samples
data = [np.random.randn() for _ in range(100_000)]

# Create subplots
fig, axs = plt.subplots(
    nrows=1,
    ncols=2,
    figsize=(10, 5),
    layout='constrained'
)

axs[0].hist(data, bins=50, color="purple")
axs[0].set_title("Histogram")
axs[0].grid(linestyle="--")

axs[1].hist(data, bins=50, cumulative=True, color="pink")
axs[1].set_title("Cumulative Histogram")
axs[1].grid(linestyle="--")

# Display figure
plt.show()
```
